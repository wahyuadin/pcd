```python
#PCD TUGAS 7
#NO 1
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from pylab import *
```

```python
img1 = cv2.imread('wahyu.jpg')
img2 = cv2.imread('wahyu_bebas.jpg')
```

```python
sift = cv2.SIFT_create()
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
keypoint1, descriptor1 = sift.detectAndCompute(img1, None)
keypoint2, descriptor2 = sift.detectAndCompute(img2, None)
sift_image1 = cv2.drawKeypoints(gray1, keypoint1, img1)
sift_image2 = cv2.drawKeypoints(gray2, keypoint2, img2)
```

```python
print('Menampilkan Fitur SIFT')
figure(0)
io.imshow(sift_image1)
figure(1)
io.imshow(sift_image2)
```

```python
#Menyimpan Hasil Citra 
#No 2
cv2.imwrite("foto_Formal.jpg", sift_image1)
cv2.imwrite("foto_Bebas.jpg", sift_image2)
```

```python
#No 3
from skimage.measure import compare_ssim as ssim

hasilformal = cv2.imread('fotoformal.jpg')
hasilbebas = cv2.imread('fotobebas.jpg')

asliFormal_hasilFormal = ssim(img1, hasilformal, data_range = hasilformal.max() - hasilformal.min(), multichannel = True)
asliBebas_hasilBebas = ssim(img2, hasilbebas, data_range = hasilbebas.max() - hasilbebas.min,multichannel = True)
hasilformal_hasilbebas = ssim(hasilformal, hasilbebas, data_range = hasilbebas.max() - hasilbebas.min, multichannel = True)


print('Perbandingan Foto asli formal dengan Hasil formal = ',asliFormal_hasilFormal)
print('Perbandingan Foto asli bebas dengan Hasil bebas = ',asliBebas_hasilBebas)
print('Perbandingan Foto hasil Formal dengan Hasil bebas = ',asliFormal_hasilbebas)
```

```python
#no 4
norm = cv2.NORM_L2
netsian = cv2.BFMatcher(norm)
matches = netsian.match(descriptor1, descriptor2)
matches = sorted(matches, key = lambda match:match.distance)
matched_img = cv2.drawMatches(img1, keypoint1, img2, keypoint2, matches[:100], img2.copy())
plt.figure(figsize=(100,300))
plt.imshow(matched_img)
```
