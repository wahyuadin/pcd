```python
import cv2
import numpy as np
from skimage import io
from skimage import exposure
from pylab import *
img = io.imread("wahyu.png")
_, img0 = cv2.threshold(img, 100, 255, 0)
_, img1 = cv2.threshold(img, 100, 255, 1)
figure(0)
io.imshow(img0)
figure(1)
io.imshow(img1)
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('wahyu.png',0)
# global thresholding
ret1,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
# Otsu's thresholding
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
# Otsu's thresholding after Gaussian filtering
blur = cv.GaussianBlur(img,(5,5),0)
ret3,th3 = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
# plot all the images and their histograms
images = [img, 0, th1,
          img, 0, th2,
          blur, 0, th3]
titles = ['Original Noisy Image','Histogram','Global Thresholding (v=127)',
          'Original Noisy Image','Histogram',"Otsu's Thresholding",
          'Gaussian filtered Image','Histogram',"Otsu's Thresholding"]
for i in range(3):
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from skimage import io
from skimage import exposure
from pylab import *
img = io.imread("wahyu.png")

cv2.GaussianBlur(img, (3, 3), 0)
konversi = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

grad_x = cv2.Sobel(konversi, cv2.CV_16S, 1, 0, ksize=3, scale=1, delta=0,
                  borderType=cv2.BORDER_DEFAULT)
grad_Y = cv2.Sobel(konversi, cv2.CV_16S, 0, 1, ksize=3, scale=1, delta=0,
                  borderType=cv2.BORDER_DEFAULT)
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv.convertScaleAbs(grad_Y)
grad = cv.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)

io.imshow(grad)
```

```python
src1 = cv2.imread("wahyu.png")
src = cv2.cvtColor(src1,cv.COLOR_BGR2GRAY)
src_hits = cv2.equalizeHist(src)
figure (0)
io.imshow(src1)
figure (1)
io.imshow(src)
figure (2)
io.imshow(src_hits)
```
