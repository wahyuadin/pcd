```python
#konversi rgb ke lab dan lch

#import library
import cv2
from skimage import io, color

#memasukan citra
img = io.imread('poto.png')
#konversi dari BGR ke RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#konversi dari RGB ke LAB
lab = color.rgb2lab(rgb)
#konversi dari LAB ke LCH
lch = color.lab2lch(lab)

cv2.imshow('RGB', rgb)
cv2.imshow('LAB', lab)
cv2.imshow('LCH', lch)
cv2.waitKey(0)
cv2.imwrite('poto_RGB.png', rgb)
cv2.imwrite('poto_LAB.png', lab)
cv2.imwrite('poto_LCH.png', lch)
cv2.destroyAllWindows()
```
